```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# داده ساختگی
torch.manual_seed(42)
x = torch.randn(100, 2)
y = (x[:, 0] + x[:, 1] > 0).float().unsqueeze(1)

# مدل لجستیک ساده
model = nn.Sequential(
    nn.Linear(2, 1),
    nn.Sigmoid()
)

# تابع هزینه (Binary Cross Entropy)
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# آموزش مدل
epochs = 100
losses = []
for epoch in range(epochs):
    y_pred = model(x)
    loss = criterion(y_pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

# نمایش نمودار
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss (Logistic Regression)')
plt.show()

# نمونه پیش‌بینی
sample = torch.tensor([[1.0, -1.0]])
print("Predicted probability:", model(sample).item())
```
